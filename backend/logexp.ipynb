```python
import numpy as np
import os

pp = os.path.join(r"D:\face\XVRFaceCap\faceTracking\VRFaceTrackingApp\backend\app\log")

exp_val = np.arange(70, 330, 10)


data = []
for id in exp_val:
    ff =os.path.join(pp, "exp_{}.txt".format(id) )
    with open(ff, 'r') as f:
        item = f.readline()
        data.append( [float(x) for x in item.split(',')] )

data = np.array(data)
data.shape

import matplotlib.pyplot as plt

# 
num_curves, num_points = len(data), len(data[0])

# X0num_points-1
x_axis = range(num_points)

plt.figure(figsize=(14, 7))
# 
for curve_data in data:
    plt.plot(x_axis, curve_data)

# 
plt.title('Curves Plot')
plt.xlabel('Size')
plt.ylabel('Value (0~1)')

# 
plt.grid(True)

# X045
plt.xticks(range(0, num_points + 1, 1))
# 
plt.show()
```
